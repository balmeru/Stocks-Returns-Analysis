```python
import pandas as pd

# List of CSV file paths
csv_file_paths = [
    
    "/Users/balmeru/Downloads/announcer_frame_1984.csv",
    "/Users/balmeru/Downloads/announcer_frame_1985.csv",
    "/Users/balmeru/Downloads/announcer_frame_1986.csv",
    "/Users/balmeru/Downloads/announcer_frame_1987.csv",
    "/Users/balmeru/Downloads/announcer_frame_1988.csv",
    "/Users/balmeru/Downloads/announcer_frame_1989.csv",

    "/Users/balmeru/Downloads/announcer_frame_1990.csv",
    "/Users/balmeru/Downloads/announcer_frame_1991.csv",
    "/Users/balmeru/Downloads/announcer_frame_1992.csv",
    "/Users/balmeru/Downloads/announcer_frame_1993.csv",
    "/Users/balmeru/Downloads/announcer_frame_1994.csv",
    "/Users/balmeru/Downloads/announcer_frame_1995.csv",
    "/Users/balmeru/Downloads/announcer_frame_1996.csv",
    "/Users/balmeru/Downloads/announcer_frame_1997.csv",
    "/Users/balmeru/Downloads/announcer_frame_1998.csv",
    "/Users/balmeru/Downloads/announcer_frame_1999.csv",
    "/Users/balmeru/Downloads/announcer_frame_2000.csv",
    "/Users/balmeru/Downloads/announcer_frame_2001.csv",
    "/Users/balmeru/Downloads/announcer_frame_2002.csv",
    "/Users/balmeru/Downloads/announcer_frame_2003.csv",
    "/Users/balmeru/Downloads/announcer_frame_2004.csv",
    "/Users/balmeru/Downloads/announcer_frame_2005.csv",
    "/Users/balmeru/Downloads/announcer_frame_2006.csv",
    "/Users/balmeru/Downloads/announcer_frame_2007.csv",
    "/Users/balmeru/Downloads/announcer_frame_2008.csv",
    "/Users/balmeru/Downloads/announcer_frame_2009.csv",
    "/Users/balmeru/Downloads/announcer_frame_2010.csv",
    "/Users/balmeru/Downloads/announcer_frame_2011.csv",
    "/Users/balmeru/Downloads/announcer_frame_2012.csv"
    
]

# Initialize an empty DataFrame to store the concatenated results
concatenated_df = pd.DataFrame()

# Loop through each CSV file path
for csv_path in csv_file_paths:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_path)
    
    # Extract the first and last columns by index
    first_column = df.iloc[:, 0]  # First column
    last_column = df.iloc[:, -1]  # Last column
    
    # Create a new DataFrame with these columns
    selected_columns_df = pd.DataFrame({
        'First_Column': first_column,
        'Last_Column': last_column
    })
    
    # Concatenate with the existing DataFrame
    concatenated_df = pd.concat([concatenated_df, selected_columns_df], ignore_index=True)

# Define the output CSV file path
output_csv_path = "/Users/balmeru/Downloads/announcer_frame.csv"

# Save the concatenated DataFrame to a CSV file
concatenated_df.to_csv(output_csv_path, index=False)

print(f"CSV file created at {output_csv_path}")
```
