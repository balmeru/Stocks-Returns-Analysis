```python
import pandas as pd
import numpy as np
from datetime import timedelta, date

# Helper function to get the Monday of the week for a given date
def get_monday(date):
    weekday = date.weekday()  # 0 for Monday, 6 for Sunday
    return date - timedelta(days=weekday) if weekday != 0 else date  # Adjust to the Monday of that week

# Helper function to get all Mondays in a given month
def get_mondays_of_month(year, month):
    mondays = []
    current_date = date(year, month, 1)  # First day of the month
    while current_date.month == month:
        if current_date.weekday() == 0:  # If it's a Monday
            mondays.append(current_date)
        current_date += timedelta(days=1)  # Move to the next day
    return mondays

# File paths for Pivot A and Pivot B
pivot_a_path = "C:/Users/balmeru/Downloads/weekly_pivot_1984.csv"
pivot_b_path = "C:/Users/balmeru/Downloads/combined_1984.csv"

# Read Pivot A and Pivot B from CSV files
pivot_a = pd.read_csv(pivot_a_path, index_col=0)  # Pivot A with Monday dates as index
pivot_b = pd.read_csv(pivot_b_path, index_col=0)  # Pivot B with report dates as index

# Ensure Pivot A's index has correct Monday dates
pivot_a.index = pd.to_datetime(pivot_a.index, errors='coerce')  # Convert to datetime

# Find common ticker symbols between Pivot A and Pivot B
common_tickers = pivot_a.columns.intersection(pivot_b.columns)

# Create Pivot C with the same index as Pivot A and common tickers
pivot_c = pd.DataFrame(index=pivot_a.index, columns=common_tickers)
pivot_c[:] = np.nan  # Initialize with NaN

# Mark True for the Mondays of report dates
for ticker in common_tickers:
    report_dates = pivot_b[ticker].dropna()  # Get all report dates for the ticker

    # Convert to datetime with format inference
    monday_dates = pd.to_datetime(report_dates, errors='coerce', format='mixed').apply(get_monday)  # Get Monday

    # Mark these Monday dates with True in Pivot C
    for monday in monday_dates:
        if monday in pivot_c.index:
            pivot_c.loc[monday, ticker] = True

    # Count the `True` values in the column (this might need to be adjusted depending on your intention)
    true_count = pivot_c[ticker].sum()

pivot_c = pivot_c.dropna(axis=1, how='all')  # Drop columns with all NaN values
# Export Pivot C to CSV
pivot_c_export_path = "C:/Users/balmeru/Downloads/True_1984.csv"  # Desired export path
pivot_c.to_csv(pivot_c_export_path)

print(f"Pivot C has been successfully saved to {pivot_c_export_path}")
```
